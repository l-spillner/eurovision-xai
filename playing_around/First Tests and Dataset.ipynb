```python
last_places = {}
for year in list(set(list(contestants["year"]))):
    last_places[year] = max(list(contestants[contestants["year"] == year]["place_contest"]))
#print(last_places)

is_last = []
for index, row in contestants.iterrows():
    if row["place_contest"] == last_places[row["year"]]:
        is_last.append(True)
    else:
        is_last.append(False)
        
#print(is_last)

contestants["is_last"] = is_last

losers = contestants[(contestants["is_last"]) & (contestants["year"]>=1990) & (contestants["year"]<=2000)]
loser_lyrics = list(losers["lyrics"])
#for l in loser_lyrics:
#    print(l)
#    print("------------------------------")
```

```python
winners_and_losers = contestants[((contestants["place_contest"] == 1) | contestants["is_last"]) & 
                                 (contestants["year"] >= 1990) & (contestants["year"] <= 2000)]
wal = winners_and_losers.to_dict('records')
random.shuffle(wal)
```

```python
results = []
```

```python
correct_w = 0
correct_l = 0
wrong_w = 0
wrong_l = 0

run_results = []

for item in wal:
    print("||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||||\n\n\n")
    print(item["song"])
    print("by", item["performer"], "from", item["to_country"])
    print("---")
    print(item["lyrics"])
    print("\n\n\n")
    prediction = input("Winner or Loser?")
    if item["is_last"]:
        print("This song LOST the Eurovision in", item["year"], ":(")
        print("Place: ", item["place_contest"])
        if prediction == "l":
            correct_l += 1
            run_results.append((item["song"], True))
        else:
            wrong_l += 1
            run_results.append((item["song"], False))
    else:
        print("This song WON the Eurovision in", item["year"], "!")
        if prediction == "w":
            correct_w += 1
            run_results.append((item["song"], True))
        else:
            wrong_w += 1
            run_results.append((item["song"], False))
    go = input("Press q to stop or any other key to continue:")
    if go == "q":
        break
    print("\n\n\n")
    
print(f"You recognized {correct_w} out of {correct_w+wrong_w} winners correctly.")
print(f"You recognized {correct_l} out of {correct_l+wrong_l} losers correctly.")
print("Accuracy:", (correct_w+correct_l)/(correct_w+correct_l+wrong_w+wrong_l))
results.append(run_results)
```
